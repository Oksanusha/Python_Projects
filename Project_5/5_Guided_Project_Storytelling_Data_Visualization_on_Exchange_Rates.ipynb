## Data Cleaning

Before we start creating data visualizations, we'll need to do a bit of data cleaning. Our focus in the guided part of the project will be on the exchange rate between the euro and the American dollar.

Below, we do the following:

-  rename the [US dollar ] and Period\Unit: columns to something easier to type — US_dollar and Time.
- change the Time column to a datetime data type.
- sort the values by Time in ascending order.
- reset the index (and drop the initial index).


```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
# drop all the rows where the '-' character appears in the US_dollar column.
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']

# convert the US_dollar column to a float data type.
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype('float')
```

```python
euro_to_dollar.info()
```

## Rolling Mean

Now that we're finished cleaning the data, we'll generate a line plot to visualize the evolution of the euro-dollar exchange rate.

```python
plt.figure(figsize=(12, 8))
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'], color='green')
plt.xlabel('Years')
plt.ylabel('US dollar')
plt.grid(True)
plt.show()
```

If we look at the line's shape, we see many small wiggles — rather than seeing a smooth line. The wiggles, however, have meaning: they are the visual representation of the daily variation in the exchange rate. The rate only shows clear upward or downward trends in the longer run (months or years).

Depending on our goals, we may not want to show that daily variation on our graph. If we want to hide it and show only the long-term trends, we can use the rolling mean.

Let's calculate the rolling means for the US_dollar column using a moving window of 30 days.

```python
euro_to_dollar['rolling_mean_30'] = euro_to_dollar['US_dollar'].rolling(30).mean()
```

```python
plt.figure(figsize=(12, 8))
plt.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean_30'], color='red')
plt.xlabel('Years')
plt.ylabel('Rolling Mean 30 days')
plt.grid(True)
plt.show()
```

Let's make plot with a various values of the rolling window: seven (one week), 30 (one month), 50 or 100 days, or 365 days (one year).

```python
plt.figure(figsize=(10,7))
plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'], color='orange')
plt.title('Original values',  weight='bold')
plt.grid(True)

for i, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
             euro_to_dollar['US_dollar'].rolling(rolling_mean).mean(), color='#04d8b2')
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')
    plt.grid(True)
    
plt.tight_layout() # Auto-adjusts the padding between subplots
plt.show()
```

## Coding the Graph


Let's show how the euro-dollar rate has changed during the coronavirus pandemic. We will show the 2020 data and the 2016-2019 data as a baseline and use a line plot.

```python
# baseline 2016-2019
corona_baseline = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2016
                   ) & (euro_to_dollar['Time'].dt.year <= 2019)]
corona_2020 = euro_to_dollar.copy(
                   )[(euro_to_dollar.Time.dt.year == 2020)]
```

```python
### Adding the FiveThirtyEight style
style.use('fivethirtyeight')

### Adding the plot
fig,ax = plt.subplots(figsize=(8,3))
ax.plot(corona_baseline['Time'],
        corona_baseline['rolling_mean_30'],
        linewidth=1, color='#0504aa')

### Highlighting the 2020 period
ax.plot(corona_2020['Time'],
        corona_2020['rolling_mean_30'],
        linewidth=3, color='#e23d28')

### Highlihting the peak of the crisis
ax.axvspan(xmin=datetime(2018, 2, 1), xmax=datetime(2018, 6, 1), ymin=0.09, alpha=0.3, color='grey')

### Adding a title and a subtitle    
ax.text(datetime(2015, 1, 1), 1.28, 'Euro-USD rate peaked at 1.25 during coronavirus pandemic',
        weight='bold')
ax.text(datetime(2015, 1, 1), 1.26, 'Euro-USD exchange rates between 2016 and 2019',
        size=12)

### Adding a signature
ax.text(datetime(2015, 1, 1), 1.0, '©OKSANUSHA' + ' '*120 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10)

### Add some transparency to the grid
ax.grid(alpha=0.5)

plt.show()
```

As we see the USD rate was getting higher in 2020 so the coronavirus pandemic really influences economic.

Now let's explore US Dollar, Russian Rouble and Japanese Yen in 2000 - 2020.

```python
exchange_rates.rename(columns={'[Russian rouble ]': 'Russian_rouble', 
                               '[Japanese yen ]': 'Japanese_yen'}, inplace=True)
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)

euro_to_multi = exchange_rates[['Time', 'US_dollar', 'Russian_rouble', 'Japanese_yen']].copy()

```

```python
euro_to_multi['Russian_rouble'].value_counts()
```

```python
euro_to_multi['Japanese_yen'].value_counts()
```

```python
euro_to_multi['US_dollar'].value_counts()
```

```python
# drop all the rows where the '-' character appears in the US_dollar column.
euro_to_multi = euro_to_multi[euro_to_multi['US_dollar'] != '-']
euro_to_multi = euro_to_multi[euro_to_multi['Russian_rouble'] != '-']
euro_to_multi = euro_to_multi[euro_to_multi['Japanese_yen'] != '-']

# convert the US_dollar column to a float data type.
euro_to_multi['US_dollar'] = euro_to_multi['US_dollar'].astype('float')
euro_to_multi['Russian_rouble'] = euro_to_multi['Russian_rouble'].astype('float')
euro_to_multi['Japanese_yen'] = euro_to_multi['Japanese_yen'].astype('float')
```

```python
euro_to_multi.info()
```

```python
# save necessary data
multi = euro_to_multi.copy(
                   )[(euro_to_multi['Time'].dt.year >= 2000) & (euro_to_multi['Time'].dt.year < 2021)]


# let's calculate rolling windows for our currencies
multi['rolling_mean_dollar'] = multi['US_dollar'].rolling(30).mean()
multi['rolling_mean_rouble'] = multi['Russian_rouble'].rolling(30).mean()
multi['rolling_mean_yen'] = multi['Japanese_yen'].rolling(30).mean()

# we must normalize 'rolling_mean_dollar', 'rolling_mean_rouble' and 'rolling_mean_yen' because these currencies have 
# different currency exchanges
first = multi['rolling_mean_dollar'] - multi['rolling_mean_dollar'].mean()
multi['normalized_dollar'] = first / first.std()
second = multi['rolling_mean_rouble'] - multi['rolling_mean_rouble'].mean()
multi['normalized_rouble'] = second / second.std()
third = multi['rolling_mean_yen'] - multi['rolling_mean_yen'].mean()
multi['normalized_yen'] = third / third.std()

multi
```

```python
### Adding the FiveThirtyEight style
style.use('fivethirtyeight')

### Adding the subplots
plt.figure(figsize=(12, 6))
ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,1,2)
axes = [ax1, ax2, ax3, ax4]
       
### Ax1: US_dollar
ax1.plot(multi['Time'], multi['rolling_mean_dollar'], color='#BF5FFF')
ax1.set_ylim(0.8, 1.7)
ax1.set_yticks([1.0, 1.2, 1.4, 1.6])
ax1.set_yticklabels(['1.0', '1.2','1.4', '1.6'], alpha=0.3)
ax1.set_xticklabels(['2000', '', '2005', '', '2010', '', '2015', '', '2020', ''], alpha=0.3)
ax1.text(datetime(1999, 1, 1), 1.78, 'US DOLLAR', fontsize=15, weight='bold', color='#BF5FFF')
ax1.grid(alpha=0.5)

### Ax2: Russian_rouble
ax2.plot(multi['Time'], multi['rolling_mean_rouble'], color='#ffa500')
ax2.set_ylim(20, 100)
ax2.set_yticks([20, 40, 60, 80, 100])
ax2.set_yticklabels(['20', '40', '60', '80', '100'], alpha=0.3)
ax2.set_xticklabels(['2000', '', '2005', '', '2010', '', '2015', '', '2020', ''], alpha=0.3)
ax2.text(datetime(1999, 1, 1), 105, 'RUSSIAN ROUBLE', fontsize=15, weight='bold', color='#ffa500')
ax2.grid(alpha=0.5)

### Ax3: Japanese_yen
ax3.plot(multi['Time'], multi['rolling_mean_yen'], color='#00B2EE')
ax3.set_ylim(90, 170)
ax3.set_yticks([80, 100, 120, 140, 160])
ax3.set_yticklabels([80, 100, 120, 140, 160], alpha=0.3)
ax3.set_xticklabels(['2000', '', '2005', '', '2010', '', '2015', '', '2020', ''],alpha=0.3)
ax3.text(datetime(1999, 1, 1), 174, 'JAPANISE YEN', fontsize=15, weight='bold', color='#00B2EE')
ax3.grid(alpha=0.5)

### Ax4: Dollar-Rouble-Yen
ax4.plot(multi['Time'], multi['normalized_dollar'], color='#BF5FFF')
ax4.plot(multi['Time'], multi['normalized_rouble'], color='#ffa500')
ax4.plot(multi['Time'], multi['normalized_yen'], color='#00B2EE')
ax4.set_ylim(-3, 3)
ax4.set_yticks([-3, -1.5, 0, 1.5, 3])
ax4.set_yticklabels([-3, -1.5, 0, 1.5, 3], alpha=0.3)
ax4.set_xticklabels(['2000', '', '2005', '', '2010', '', '2015', '', '2020', ''], alpha=0.3)
ax4.grid(alpha=0.5)

### Adding a title and a subtitle
ax1.text(datetime(1999, 1, 1), 2.0, 'EURO-USD-ROUBLE-YEN rate between 2000 and 2020',
         fontsize=20, weight='bold')

### Adding a signature
ax4.text(datetime(1999, 1, 1), -4.5, '©OKSANUSHA' + ' '*103 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=14)

plt.show()

```

During the financial crisis of 2008, we see that the rouble was stable relative to the euro, but rose sharply in 2009. In turn, the yen and the dollar rose against the euro in 2008. As of 2020, the rouble has a higher exchange rate against the euro against the dollar and the yen. Most likely, this is due to political reasons.
